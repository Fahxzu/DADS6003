```python
#avg_income.py
from pyspark import SparkContext

spark_assign = SparkContext(appName='avg_income')
input_file = spark_assign.textFile('/user/cloudera/sw/avg_income_test.txt')
tokens = input_file.map(lambda line : line.split(','))


curravg = tokens.reduceByKey(lambda x,y : (int(x)+int(y))/2)
curravg.saveAsTextFile('/user/cloudera/sw/output')
spark_assign.stop()
```

```python
from pyspark import SparkContext, SparkConf

def main():
  conf = SparkConf().setAppName("avg_Income")
  sc = SparkContext(conf=conf)
    
  lines = sc.textFile("/user/cloudera/test.txt")

  parsed_lines = lines.map(lambda x: tuple(x.split(",")))

  income_by_continent = parsed_lines.map(lambda x: (x[1], (1, int(x[2]))))

  total_income_by_continent = income_by_continent.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))

  average_income_by_continent = total_income_by_continent.mapValues(lambda x: x[1] / x[0])
    
  average_income_by_continent.saveAsTextFile('/user/cloudera/sw/assignt_output')

  results = average_income_by_continent.collect()
  for result in results:
    print(result)
    
  sc.stop()

'''if __name__ == "__main__" : 
  main()'''
```
