```python
#reducer.py
#!/usr/bin/env python
import sys

current_conti = None
total_income = 0
country_num = 0

for i in sys.stdin:
    i = i.strip()
    continent, income = line.split('\t', 1)

    try : income = int(income)
    except ValueError : continue

    if current_conti == continent:
        total_income += income
        country_num += 1

    elif current_conti:
      average_income = total_income / country_num
      print('%s\t%s' % (current_conti, avg_income))
    current_conti = continent
    total_income = income
    country_num = 1

# print the avg income for the last continent
if current_conti == 'continent':
    avg_income = total_income / country_num
    print('%s\t%s' % (current_conti, avg_income))
```
